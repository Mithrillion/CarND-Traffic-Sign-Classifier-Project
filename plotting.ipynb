```python
# %load /home/mithrillion/default_imports.py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 8)
%matplotlib inline
import re
from collections import Counter
from functools import partial
```

```python
train = pd.read_csv("./summary/run_.-tag-train_loss.csv")
val = pd.read_csv("./summary/run_.-tag-validation_loss.csv")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(train['Step'], train['Value'], color='r')
plt.plot(val['Step'], val['Value'], color='g')
plt.legend(['training', 'validation'])
plt.title("training loss vs validation loss")
plt.xlabel("step")
plt.ylabel("cross-entropy loss")
plt.show()
```
